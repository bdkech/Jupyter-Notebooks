```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

data = pd.read_csv(open('prot_mrna_cnv_data.txt','r'),sep="\t")
```

```python
for gene_name in data['Gene'].unique():
    to_process = data[data['Gene'] == gene_name]
    try:
        model      = ols("prot_z ~ mrna_z + SEG_MEAN",
                         data=to_process).fit()
        print model.rsquared**.5
    except:
        print "NA"
```
