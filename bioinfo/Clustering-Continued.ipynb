### Metadata and cluster information
#### Below is the metadata and cluster assignment for the genes with most information from the 3 Component PCA

```python
clustered_data = pd.read_csv(open("mrna_with_cluster_spearman.txt",'r'),delimiter="\t")
clustered_data.head()
d = widgets.Select()
d.options = clustered_data.columns.tolist()

def plot_hist(filter_val):
    clustered_data.groupby(['cluster_id',filter_val]).size().plot(kind="bar",figsize=(12,4))
    
widgets.interactive(plot_hist,filter_val=d)
```

#### This is the metadata and cluster mapping for all 50 genes 3 component PCA

```python
clustered_data2 = pd.read_csv(open("mrna_with_cluster_all_50_pca_3_spearman.txt",'r'),delimiter="\t")
d2 = widgets.Select()
d2.options = clustered_data2.columns.tolist()

def plot_hist2(filter_val2):
    clustered_data2.groupby(['cluster_id',filter_val2]).size().plot(kind="bar",figsize=(12,4))
    
widgets.interactive(plot_hist2,filter_val2=d2)
```

#### Below are the 2 component PCAs, first the most informative genes, then all 50.
It is worth noting, that for the most informative genes there wasn't much of a change from 3 to 2.  For all 50 genes, there seemed to be the generation of an additional cluster that has only one member.  The overall message is that with the 50 gene model, we essentially get two groupings that have the occurrance of metastasis.
